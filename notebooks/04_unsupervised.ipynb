Clusters origin airports by traffic and delay behavior.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT / 'src'))
```

```python
from flight_delays.data import load_flights
from flight_delays.unsupervised import build_airport_profiles, cluster_airport_profiles, describe_clusters
from flight_delays.reporting import plot_clusters

flights = load_flights(nrows=100_000)
profiles = build_airport_profiles(flights, min_flights=50)
profiles.head()
```

```python
clustered_profiles, cluster_model = cluster_airport_profiles(profiles, n_clusters=4)
clustered_profiles.head()
```

```python
cluster_summary = describe_clusters(clustered_profiles)
cluster_summary
```

```python
plot_clusters(clustered_profiles)
```
